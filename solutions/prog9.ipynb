```python
input_fname = "../inputs/day9.in"
```

```python
a_moves = []
with open(input_fname) as ifile:
    def ri(): 
        return int(ifile.readline().strip())
    def ris(): 
        return list(map(int, ifile.readline().strip().split()))
    for line in ifile:
        d, n = line.strip().split()
        a_moves.append((d, int(n)))
a_moves
```

```python
delta = dict(R=1, L=-1, U=1j, D=-1j)
```

```python
h = 0+0j
ts = [0+0j for _ in range(9)]
visited = {ts[-1]}
for d, n in a_moves:
    print("\n====", d, n, "====")
    for _ in range(n):
        h += delta[d]
        # print("initial", h, t)
        print("h", h)
        follow = h
        for i,ti in enum(ts):
            oldt = ti
            diff = follow-ti
            if abs(diff.imag) <= 1 and abs(diff.real) <= 1:
                print("\t", i, ti, "(no change)")
            else:
                reals, imags = np.sign(diff.real), np.sign(diff.imag)
                ti += min(1, abs(diff.real))*reals + min(1, abs(diff.imag))*imags*1j
                print("\t", i, oldt, "->", ti)
                ts[i] = ti
            follow = ti
        visited.add(ts[-1])
```
